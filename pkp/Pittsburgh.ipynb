```python
from scipy.optimize import brute, minimize, fmin, basinhopping

def generate_fitted_yield(cpd, spec):
    pyrol_yield = Yield(cpd, spec)
    cons = ({'type': 'eq', 'fun': lambda x:  1 - sum(x)})
    init = [0.0] * len(spec)
    optimizer = lambda: minimize(pyrol_yield.error_func, init, tol=0.001, bounds=[(0, 1.0) for _ in init], constraints=cons).x
    return pyrol_yield.fit_composition(optimizer), pyrol_yield


def generate_fitted_yield_manual(cpd, q, spec):
    pyrol_yield = Yield(cpd.coal, spec)
    cons = ({'type': 'eq', 'fun': lambda x:  1 - sum(x)})
    init = [0.0] * len(spec)
    optimizer = lambda: minimize(pyrol_yield.error_func, init, tol=0.001, bounds=[(0, 1.0) for _ in init], constraints=cons).x
    return pyrol_yield.fit_composition(optimizer), pyrol_yield

from pkp.src.CoalThermoPhysics import enthalpy_balance

def enthal_of_devol(compositions_):
    dT = {
        #'HHV (as received) [kJ/kg_coal]': float(y.pre.coal.hhv),
        #'HHV (as received,Dulong) [kJ/kg_coal]': float(y.pre.coal.hhv_dulong*1000.0),
        'HHV (daf) [kJ/kg_coal_daf]': [float(comp.hhv_daf) for fit, comp in compositions_.values()],
        'LHV (daf) [kJ/kg_coal_daf]': [float(comp.lhv_daf) for fit, comp in compositions_.values()],
        'LHV (yield) [kJ/kg_yield]': [float(comp.lhv_yield_species) for fit, comp in compositions_.values()],
        'h_devol [kJ/kg_yield]': [float(comp.EnthalpyOfDevol) for fit, comp in compositions_.values()]
    }
    names = compositions_.keys()
    return DictTable(dT, name=names, caption="Enthalpy Balances")
```

```python
def plt_bar(f, ax ,a, label_, pattern, offs=0.0):
    bar = 0.50
    ua = zip(*[(label, value) for label, value in a.iteritems()])
    index = np.arange(len(ua[0]))
    ax.bar(index+offs, ua[1], bar, label=label_, hatch=pattern, color="None")
    ax.set_ylabel('Composition [%]')
    ax.get_xaxis().set_ticklabels(ua[0])
    ax.get_xaxis().set_ticks([0.5, 1.5, 2.5, 3.5])
    plt.legend(loc="upper left")
```

```python
from pkp.src.Models import arrheniusRate

inp = {'arrheniusRate': {
            'preExp': 0.0, 'preExpBounds': [100.0, 100000.0],
            'beta':0.0, 'activationEnergy': 0,
            'betaBounds':(0.0, 0.0), 'lowerDevolTemp': 10.0,
            'activationEnergyBounds':[100.0, 10000.0]
            }
       }
inp_michele = deepcopy(inp)
inp_michele['arrheniusRate']['lowerDevolTemp'] = 100.0

arrhRate_pittsburgh = arrheniusRate(inp, {'run0':cpd_pittsburgh.values()[0]}, 'ftot')
x = arrhRate_pittsburgh.fit(delta=0.05, finalOpt=False)
yieldArrRate_pittsburgh = arrhRate_pittsburgh.fittedYield(cpd_pittsburgh.values()[0])#
```
