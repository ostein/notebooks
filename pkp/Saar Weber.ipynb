```python
# Based on IFRF DOC No F36/y/20

input_dict = {'Coal': {
        'Proximate Analysis': {"Fixed Carbon": 52.5, "Moisture": 2.0,
                               "Volatile Matter": 37.0,  "Ash": 8.3},
        'Ultimate Analysis': {"Carbon": 79.3, "Hydrogen": 4.7,
                              "Nitrogen": 1.3, "Oxygen": 13.7,
                              "Sulphur": 1.0},
        'hhv': 27000, 'MW_PS': 100},
    'CPD': {'fit': 'constantRate', 'active': 'True',
              'deltaT': '0.5e-05', 'MW_TAR': '130'}
    , 'OperatingConditions': {'pressure': 1.0}}

side(
    DictTable(input_dict['Coal']['Ultimate Analysis'], caption="Ultimate Analysis"),
    DictTable(input_dict['Coal']['Proximate Analysis'], caption="Proximate Analysis")
)
```

# Compute Particle Temperatures

```python
from pkp.src.Models import ParticleHeatUp

for i, particle_diam in enumerate([25e-6 ,50e-6, 75e-6, 100e-6, 150e-6]):
    particleHeatUpModel = ParticleHeatUp(
        cp=1000.0, rho=1000.0, d=particle_diam,
        sigma=5.6e-8, epsilon=0.8, T_wall=1500.0,)
    times, temps = particleHeatUpModel.computeParticleTemperature([0.0, 0.05],[1500.0, 2500.0], t_end=0.035)
    dict_ = {'run'+str(i): zip(times, temps)[::200]}
    input_dict['OperatingConditions'].update(dict_)
```

```python
cpd_pittsburgh = pkp.generate(json_string=input_dict)
cpd_pittsburgh.timesteps = 15
```

```python
from pkp.src.CoalThermoPhysics import enthalpy_balance

def enthal_of_devol(compositions_):
    dT = {
        #'HHV (as received) [kJ/kg_coal]': float(y.pre.coal.hhv),
        #'HHV (as received,Dulong) [kJ/kg_coal]': float(y.pre.coal.hhv_dulong*1000.0),
        'HHV (daf) [kJ/kg_coal_daf]': [float(comp.hhv_daf) for fit, comp in compositions_.values()],
        'LHV (daf) [kJ/kg_coal_daf]': [float(comp.lhv_daf) for fit, comp in compositions_.values()],
        'LHV (yield) [kJ/kg_yield]': [float(comp.lhv_yield_species) for fit, comp in compositions_.values()],
        'h_devol [kJ/kg_yield]': [float(comp.EnthalpyOfDevol) for fit, comp in compositions_.values()]
    }
    names = compositions_.keys()
    return DictTable(dT, name=names, caption="Enthalpy Balances")
```
