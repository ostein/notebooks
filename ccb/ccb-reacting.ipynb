```python
from functools import partial
from functools import reduce as freduce
from functional import compose

multi_compose = partial(freduce, compose)

rowLabel = partial(ow.plot.adjustColumn,{'yaxis.visible': False}, ow.plot.all, ow.plot.tail)
size = partial(ow.plot.adjustColumn,{'min_border':12, 'plot_height': 300, 'plot_width': 400}, ow.plot.all, ow.plot.all)
cleanTitle = partial(ow.plot.adjustColumn, {'title':""}, ow.plot.tail, ow.plot.all)
cleanAxis = partial(ow.plot.adjustColumn, {'xaxis.visible': False}, ow.plot.rtail,ow.plot.all)

style = multi_compose([rowLabel, size, cleanTitle, cleanAxis])

def cent_mean(sets, values, locs=["centreLine"]):
    bk.figure()
    rows=[]
    for i, v in enumerate(values):
        legend = (True if i==0 else False)
        figs = ow.plot_cases(sets, y=v, order=locs, legend=legend)
        rows.append(figs)
    bk.show(bk.GridPlot(children=rows))

def rad_mean(sets, values, locs):
    bk.figure()
    rows=[]
    for i, v in enumerate(values):
        legend = (True if i==0 else False)
        figs = ow.plot_cases(sets, y=v, order=locs, legend=legend)
        figs = [figs[1], figs[2], figs[0]]
        rows.append(figs)
    bk.show(bk.GridPlot(children=style(rows=rows)))

def over_time(sets, *fields):
    plots = {}
    for f in fields:
        cur_plot=[]
        for loc in locs:
            figure=(sets.location(loc)
                .by_index('Time')
                .show(f, overlay=True, legend=False, title=loc))
            cur_plot.append(figure)
        plots[f]=cur_plot
    bk.show(
        bk.GridPlot(
            children=style(list(plots.values())))
    )
```
