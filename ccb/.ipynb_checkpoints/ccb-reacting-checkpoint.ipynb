```python
from bokeh.models import (
    BasicTicker, ColumnDataSource, Grid, GridPlot, LinearAxis,
    DataRange1d, PanTool, Plot, WheelZoomTool
)

from functools import partial
from functional import compose

rowLabel = partial(ow.plot.adjustColumn,{'yaxis.visible': False}, ow.plot.all, ow.plot.tail)
size = partial(ow.plot.adjustColumn,{'min_border':12, 'plot_height': 300, 'plot_width': 400}, ow.plot.all)
cleanTitle = partial(ow.plot.adjustColumn, {'title':""}, ow.plot.tail)
cleanAxis = partial(ow.plot.adjustColumn, {'xaxis.visible': False}, ow.plot.rtail)

style = compose(rowLabel, size)


def cent_mean(sets, values, locs=["centreLine"]):
    bk.figure()
    rows=[]
    for i, v in enumerate(values):
        legend = (True if i==0 else False)
        figs = ow.plot_cases(sets, y=v, order=locs, legend=legend)
        rows.append(figs)
    bk.show(bk.GridPlot(children=rows))

def rad_mean(sets, values, locs):
    bk.figure()
    rows=[]
    for i, v in enumerate(values):
        legend = (True if i==0 else False)
        figs = ow.plot_cases(sets, y=v, order=locs, legend=legend)
        figs = style([figs[1], figs[2], figs[0]])
        rows.append(figs)
    bk.show(bk.GridPlot(children=(rows)))
    

def cleanLabelColumn(rows):
    for i, row in enumerate(rows[1:]):
        for j,fig in enumerate(row):
            rows[i][j].xaxis.axis_label=""
    for i, row in enumerate(rows[1:]):
        for j,fig in enumerate(row):
            rows[i+1][j].title=""
    return rows

def over_time(sets):
    fields = ['wMean', 'uMean', 'TMean']
    plots = {}
    for f in fields:
        cur_plot=[]
        for loc in locs:
            figure=(sets.location(loc)
                .by_index('Time')
                .show(f, overlay=True, legend=False, title=loc))
            cur_plot.append(figure)
        plots[f]=cur_plot
    bk.show(
        bk.GridPlot(
            children=style(*list(plots.values())))
    )
```

```python
over_time(cases["c1=4.0c2=0.175-CxHyOz"])
```
